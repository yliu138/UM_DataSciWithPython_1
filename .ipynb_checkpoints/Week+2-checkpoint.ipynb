df.rename(
    ['mapper=None', 'index=None', 'columns=None', 'axis=None', 'copy=True', 'inplace=False', 'level=None'],

```python
# It will iterate through the columns directly on the df object
# inplace=True is very important
for col in df.columns:
    print(col, col[:2])
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```

# Querying a DataFrame
1. Boolean masking is the heart of fast and efficient querying in NumPy.


```python
df.head()
```

```python
df['Gold'] > 0
```

```python
# Most statistical functions built into the DataFrame object ignore values of NaN.
# Notice that df['Gold'] > 0 will return a full list with boolean values, which is called boolean masking
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

```python
# the nan will be ignored by dataframe functions
only_gold['Gold'].count()
```

```python
df['Gold'].count()
```

```python
only_gold = only_gold.dropna()
only_gold.head()
```

Here's a more concise example of how we could query this DataFrame. You'll notice that there are no NaNs when you query the DataFrame in this manner. pandas automatically filters out the rows with now values.

```python
only_gold = df[df['Gold'] > 0]
only_gold.head()
```

```python
len(df[(df['Gold'] > 0) | (df['Gold.1'] > 0)])
```

```python
df[(df['Gold.1'] > 0) & (df['Gold'] == 0)]
```
